```python
# Import Necessary Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

<h4>Load the Data<h4>

```python
data = pd.read_csv("insurance.csv")
```

Let's take a look at what our data looks like

```python
data.sample(5)
```

<h5>We have 7 features:</h5>

    age: age of primary beneficiary

    sex: insurance contractor gender, female, male

    bmi: Body mass index, providing an understanding of body, weights that are relatively high or low relative to height,
    objective index of body weight (kg / m ^ 2) using the ratio of height to weight, ideally 18.5 to 24.9

    children: Number of children covered by health insurance / Number of dependents

    smoker: Smoking

    region: the beneficiary's residential area in the US, northeast, southeast, southwest, northwest.

    charges: Individual medical costs billed by health insurance
<h5>We have one interesting feature,which is BMI. It's interesting,because we can either leave it as numeric value or we could make it categorical, by referring to American Cancer Society, which tells us we can have 4 ranges for BMI: Underweight, Normal Weight, Overweight and Obese. We will think about doing it after Exploratory Data Analysis.</h5>
<h6>Note: BMI Index is the same for adult males and adult females</h6>

```python
# Checking for missing values
data.isnull().sum()
```

Looks like we don't have missing values. Now let's graph the dataset for visual exploration.

```python
# First we will start with basic describe method
data.describe()
```

From this we can get an idea of what kind of numbers we are dealing with and just general overview of the dataset.


```python
# Now let's visualize dataset with a heatmap

plt.figure(figsize=(10,5))
sns.heatmap(data=data.select_dtypes(include=[np.number]).corr(),annot=True)
```

<h6>As we can see we have some positive correlation between age and charges,as well as bmi and charges. This means as either of the values rises,the other one will rise as well. Let's look at the pairplot.</h6>


```python
sns.pairplot(data=data)
```

<p>Just as a confirmation, we can see that for lower BMI scores charges are lower. Also, we can see prices tilting to the right as age increases.</p>
<p>Also, from this pairplot we can see that the number of children covered by insurance doesn't really impact the price, although we can see starting from 4 children - prices do go down.</p>
<h5>Now let's plot the distribution of charges</h5>

```python
plt.figure(figsize=(12,8))
plt.xlim(0,64000)
sns.histplot(data=data, x="charges", hue="sex")
```

<h5>So, most of the insurance prices lies in the pre 30000 dollars with the majority lying under 15000 mark. Also, we can see that men have slightly more counts of paying the least for insurance,but overall insurance is cheaper for women.</h5>
<p> Now let's see the same distribution but for smokers vs non-smokers </p>

```python
plt.figure(figsize=(12,8))
plt.xlim(0,64000)
sns.histplot(data=data, x="charges", hue="smoker")
```

<h5>This is a very strong correlation between prices of insurance and being a smoker. As we can see, for smokers prices START at 15000 dollars mark, as lower prices are almost completely occupied by non-smokers. </h5>
<p> Something I want to explore is how much smokers there are between men and women </p>

```python
sns.histplot(data=data, x="smoker", hue="sex")
```

As I was expecting, men are smokers more often than women, but the difference is not astounding. Although it could have an impact on women having cheaper insurance overall.
<h5> Let's quickly look once more on how prices go up with age. </h5>

```python
plt.figure(figsize=(12,5))
sns.scatterplot(data=data, x="charges", y="age", hue="age")
```

This is a coloured plot from the pairplot we saw earlier. Here we can more distinctly see how age impacts charges.

<h5> Lastly I want to explore if region has anything to do with the price of insurance </h5>

```python
plt.figure(figsize=(12,5))
sns.barplot(data=data, y="charges", x="region")
```

It looks like the region does indeed impact the charges. Now let's look at the same barplot but for men and women.

```python
plt.figure(figsize=(12,5))
sns.barplot(data=data, y="charges", x="region", hue="sex")
```

This barplot shows something interesting. It seems like in all regions men pay more for insurance,which we have already discovered, but it looks like in the northwest region women pay on average slightly more than men.

<h3>Conclusion</h3>
<p>It looks like every feature has smaller or bigger impact on the charges, with Smoker feature being the most impactful.</p>
<p>There is indeed difference between prices for men and women. Women tend to pay less for insurance.</p>
<p>Answering the question of whether age impacts price of insurance, we can definitely say that there is a correlation. Insurance charges tend to go up as you get older.</p>
<h5>Additional takeaways</h5>
<p>People with smaller BMI Index pay less for insurance.</p>
<p>People with 4 or more children also pay less for insurance.</p>
<p>Although women on average pay less than men, in the northwest region of USA both genders pay almost equally, with women paying slighly more.</p>

<h2> Data preprocessing</h2>

After EDA I decided to leave BMI as it is because it has shown impact on the charges the way it already is. 
<h5>Dealing with Categorical Variables.</h5>

```python
data.select_dtypes(include=(object)).columns
```

Creating dummy variables for categorical features.

```python
dummies = pd.get_dummies(data[['sex', 'smoker','region']], drop_first=True)
data = pd.concat([data.drop(['sex', 'smoker','region'],axis=1),dummies],axis=1)
```

Let's look how dataset looks now.

```python
data.sample(3)
```

Now let's Train Test Split.

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop('charges',axis=1)
y = data['charges']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

Now let's scale our data.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```
